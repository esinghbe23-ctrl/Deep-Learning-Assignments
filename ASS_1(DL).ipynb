```python
import numpy as np
from sklearn.metrics import accuracy_score
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

T = 1.0
F = 0.0
def get_OR_data():
    X = [
        [F, F],
        [F, T],
        [T, F],
        [T, T]
    ]

    Y = [
        [F],
        [T],
        [T],
        [T]
    ]

    return np.array(X), np.array(Y)
```

```python
class MyPerceptron:

    def __init__(self, learning_rate=0.1, n_iterations=1000):
        self.lr = learning_rate
        self.epochs = n_iterations
        self.weights = None
        self.bias = None

    def step_activ_func(self, activation):
        if activation >= 0:
            return 1
        else:
            return 0

    def fit(self, X, Y):

        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for epoch in range(self.epochs):
            for i in range(X.shape[0]):

                linear_output = np.dot(self.weights, X[i]) + self.bias
                Y_pred = self.step_activ_func(linear_output)

                error = Y[i] - Y_pred

                self.weights = self.weights + self.lr * error * X[i]
                self.bias = self.bias + self.lr * error

    def predict(self, X):

        predictions = []

        for i in range(X.shape[0]):
            linear_output = np.dot(self.weights, X[i]) + self.bias
            predictions.append(self.step_activ_func(linear_output))

        return np.array(predictions)

```

```python
X, Y = get_OR_data()

clf = MyPerceptron()
clf.fit(X, Y)

X_test = np.array([
    [F, F],
    [T, F],
    [F, T],
    [T, T]
])

Y_test = np.array([
    0,
    1,
    1,
    1
])

Y_predicted = clf.predict(X_test)

print("Perceptron Predictions:", Y_predicted)
print("Accuracy:", accuracy_score(Y_test, Y_predicted))

```

```python
def get_XOR_data():

    X = [
        [F, F],
        [F, T],
        [T, F],
        [T, T]
    ]

    Y = [
        [F],
        [T],
        [T],
        [F]
    ]

    return np.array(X), np.array(Y)
```

```python
X, Y = get_XOR_data()

X_test = np.array([
    [T, F],
    [T, T],
    [F, T],
    [F, F]
])

Y_test = np.array([
    [T],
    [F],
    [T],
    [F]
])
```

```python
model = Sequential()

model.add(Dense(16, input_shape=(2,), activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(
    loss='binary_crossentropy',
    optimizer='SGD',
    metrics=['accuracy']
)

model.fit(X, Y, epochs=100, verbose=0)

loss, accuracy = model.evaluate(X_test, Y_test, verbose=0)

print("NN Accuracy: %.2f%%" % (accuracy * 100))
print("Loss:", loss)
```
